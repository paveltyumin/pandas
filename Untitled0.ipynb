```python
import pandas as pd
import random

lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI': lst})

# Создаем кросстабуляцию между столбцами 'whoAmI' и фиктивным столбцом
crosstab_result = pd.crosstab(index=data['whoAmI'], columns='count')

# Объединяем результат кросстабуляции с исходным DataFrame
data_one_hot = pd.merge(data, crosstab_result, left_on='whoAmI', right_index=True)

# Удаляем столбец 'whoAmI' и переименовываем столбец 'count' в 'robot' и 'human'
data_one_hot = data_one_hot.drop(columns='whoAmI')
data_one_hot = data_one_hot.rename(columns={'count_robot': 'robot', 'count_human': 'human'})

data_one_hot.head()








```
